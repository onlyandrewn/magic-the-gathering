# ALL SETS

### Average number of cards in all sets

```python
df = pd.read_json("https://api.scryfall.com/sets")
df["data"][0]
```

```python
df = pd.read_json("https://api.scryfall.com/sets")

all_sets = []

for set in df["data"]:
    year = int(set["released_at"].split("-")[0])

    if year >= 1993 and year <= 1998: 
        group = "1993-1998"
    elif year >= 1999 and year <= 2004:
        group = "1999-2004"
    elif year >= 2005 and year <= 2010:
        group = "2005-2010"  
    elif year >= 2011 and year <= 2016:
        group = "2011-2016"
    elif year >= 2017 and year <= 2022:
        group = "2017-2022"
        
    if year >= 1993 and year <= 2002: 
        decade = "1993-2002"
    elif year >= 2003 and year <= 2012:
        decade = "2003-2012"
    elif year >= 2013 and year <= 2022:
        decade = "2013-2022"  

    info = {
        "name": set["name"],
        "scryfall_uri": set["scryfall_uri"],
        "released_at": set["released_at"],
        "released_year": int(set["released_at"].split("-")[0]),
        "released_month": int(set["released_at"].split("-")[1]),
        "released_day": int(set["released_at"].split("-")[2]),
        "card_count": set["card_count"],
        "set_type": set["set_type"],
        "group": group,
        "decade": decade
    }

    all_sets.append(info)
```

```python
df = pd.DataFrame(all_sets)
df
```

```python
df[df["released_year"] >= 2020].head(50)
```

```python
df.to_csv("all_sets.csv", index=False)
```

# EXPANSION OR CORE ONLY

### Average number cards in a core or expansion product

```python
df = pd.read_json("https://api.scryfall.com/sets")

sets = []

for set in df["data"]:
    if set["set_type"] == "expansion" or set["set_type"] == "core":
        year = int(set["released_at"].split("-")[0])

        if year >= 1993 and year <= 1998: 
            group = "1993-1998"
        elif year >= 1999 and year <= 2004:
            group = "1999-2004"
        elif year >= 2005 and year <= 2010:
            group = "2005-2010"  
        elif year >= 2011 and year <= 2016:
            group = "2011-2016"
        elif year >= 2017 and year <= 2022:
            group = "2017-2022"

        if year >= 1993 and year <= 2002: 
            decade = "1993-2002"
        elif year >= 2003 and year <= 2012:
            decade = "2003-2012"
        elif year >= 2013 and year <= 2022:
            decade = "2013-2022"  

        info = {
            "name": set["name"],
            "scryfall_uri": set["scryfall_uri"],
            "released_at": set["released_at"],
            "released_year": int(set["released_at"].split("-")[0]),
            "released_month": int(set["released_at"].split("-")[1]),
            "released_day": int(set["released_at"].split("-")[2]),
            "card_count": set["card_count"],
            "set_type": set["set_type"],
            "group": group,
            "decade": decade
        }

        sets.append(info)
```

```python
df = pd.DataFrame(sets)
df
```
